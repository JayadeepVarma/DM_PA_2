```python
import os
import numpy as np
from PIL import Image
from skimage import io
from skimage.color import rgb2gray
from skimage import filters
from skimage import exposure, img_as_float
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
%matplotlib inline
import json
import warnings
warnings.filterwarnings("ignore")

main_dir = r'D:\Data Mining\Programming Assignment - 1\Codes\Cropped-1'
class_dir = [os.path.join(main_dir, class_name)
              for class_name in os.listdir(main_dir) 
              if os.path.isdir(os.path.join(main_dir, class_name))]
print("The four classes in my dataset are below\n")
def get_images():
    all_images = []
    for c in class_dir:
        files = [os.path.join(c, file)
                 for file in os.listdir(c)
                 if file.endswith(('.jpg'))]
        print(f"The {os.path.basename(c)} has {len(files)} images.")
        all_images.extend(files)
    return all_images

image_files = get_images()
print(f"\nThe total number of images are {len(image_files)}.")
i=0
edge_histograms = []
reduced_edge_histograms = []

def angle(dx, dy):
        return np.mod(np.arctan2(dy, dx), np.pi)

for file in image_files:
    # GrayScale
    original = io.imread(file)
    grayscale = rgb2gray(original)
    # Angle as the direction of edge gradient at the pixel
    angle_sobel = angle(filters.sobel_h(grayscale),  filters.sobel_v(grayscale))
    # Histogram
    hist_data=exposure.histogram(angle_sobel, nbins=36)
    edge_histograms.append(hist_data)
    hist_counts, bin_edges = hist_data
    # Normalized Histogram
    hist_counts_norm = hist_counts / hist_counts.sum()
    #print(f"\nHistogram Data {i+1}\n {hist_counts_norm}")
    i+=1
    reduced_edge_histograms.append(hist_counts_norm)
```

```python
from sklearn.model_selection import train_test_split

def get_images_by_class():
    class_images = {}
    for c in class_dir:
        files = [os.path.join(c, file) for file in os.listdir(c) if file.endswith('.jpg')]
        class_name = os.path.basename(c)
        class_images[class_name] = files
    return class_images

# Get images organized by class
class_images = get_images_by_class()

# Prepare lists to hold training and test datasets
train_files = []
test_files = []

# Split each class into training and testing sets
for class_name, files in class_images.items():
    train, test = train_test_split(files, test_size=0.2, random_state=42)
    train_files.extend(train)
    test_files.extend(test)
    print(f"Class '{class_name}': {len(train)} training images, {len(test)} test images")

print(f"\nTotal training images: {len(train_files)}")
print(f"Total testing images: {len(test_files)}")
```

```python
import os
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_predict
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Step 1: Define main directory and get images by class
main_dir = r'D:\Data Mining\Programming Assignment - 1\Codes\Cropped-1'
class_dir = [os.path.join(main_dir, class_name)
             for class_name in os.listdir(main_dir) 
             if os.path.isdir(os.path.join(main_dir, class_name))]

def get_images_by_class():
    class_images = {}
    for c in class_dir:
        files = [os.path.join(c, file) for file in os.listdir(c) if file.endswith('.jpg')]
        class_name = os.path.basename(c)
        class_images[class_name] = files
    return class_images

# Organize images by class
class_images = get_images_by_class()

# Step 2: Split each class into training and testing sets
train_files = []
test_files = []

for class_name, files in class_images.items():
    train, test = train_test_split(files, test_size=0.2, random_state=42)
    train_files.extend(train)
    test_files.extend(test)

# Step 3: Create train_labels for each image in train_files
train_labels = []
file_to_label = {file: class_name for class_name, files in class_images.items() for file in files}
train_labels = [file_to_label[file] for file in train_files]

# Step 4: Load histograms and apply standardization
# Assume `reduced_edge_histograms` is a list of precomputed histograms for each image

# Prepare training histograms
train_histograms = np.array([reduced_edge_histograms[i] for i, file in enumerate(train_files)])
scaler = StandardScaler()
train_histograms_scaled = scaler.fit_transform(train_histograms)

# Step 5: Initialize classifiers for cross-validation
classifiers = {
    'Naive Bayes': GaussianNB(),
    'Neural Network': MLPClassifier(hidden_layer_sizes=(10, 10, 10), max_iter=500),
    'Adaboost': AdaBoostClassifier()
}

# Step 6: Create a 5-fold stratified splitter
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Dictionary to store confusion matrices
conf_matrices = {}

# Perform cross-validation
for clf_name, clf in classifiers.items():
    # Perform cross-validated predictions
    y_pred = cross_val_predict(clf, train_histograms_scaled, train_labels, cv=skf)
    
    # Generate confusion matrix
    conf_matrices[clf_name] = confusion_matrix(train_labels, y_pred)

# Step 7: Plot confusion matrices one after the other (vertically)
fig, axes = plt.subplots(3, 1, figsize=(10, 24))  # Adjust figsize for vertical layout

for ax, (clf_name, cm) in zip(axes, conf_matrices.items()):
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(train_labels))
    im = disp.plot(ax=ax, cmap="viridis", values_format='d', colorbar=False)  # Set colorbar=False to avoid auto color bar
    
    ax.set_title(f"Confusion Matrix for {clf_name}", fontsize=16)
    ax.set_xlabel("Predicted label", fontsize=14)
    ax.set_ylabel("True label", fontsize=14)
    
    # Rotate x-axis labels for vertical orientation and adjust alignment
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha='center', va='top', fontsize=12)
    ax.tick_params(axis='both', which='major', labelsize=12)
    
    # Manually add a color bar with padding
    fig.colorbar(im.im_, ax=ax, fraction=0.046, pad=0.04)

# Adjust the overall layout to add more space on the right side
plt.subplots_adjust(right=0.85)  # Move this value closer to 1 for more space

plt.tight_layout()
plt.show()
```

After visually comparing the confusion matrices of three modelsNaive Bayes, Neural Network, and AdaBoostAdaBoost emerged as the best-performing method. This conclusion is based on the following observations:

Higher Diagonal Values: AdaBoost has the most accurate predictions, as seen by the higher values along the diagonal. This indicates it correctly classifies more instances in each class compared to the other models.
Fewer Misclassifications: AdaBoost has fewer off-diagonal values, meaning it makes fewer errors in classifying instances across the classes.
In contrast, Naive Bayes showed significant misclassification rates, likely due to its assumptions not aligning well with the data distribution. The Neural Network performed moderately, but still had more misclassifications than AdaBoost.

Overall, AdaBoost's superior diagonal values and lower misclassification rates make it the most reliable model among the three for this dataset.

```python
import os
import random
import numpy as np
from sklearn.svm import LinearSVC
from sklearn.model_selection import cross_val_score, StratifiedKFold, KFold
from sklearn.preprocessing import StandardScaler

# Define the directory path for the images
main_dir = r'D:\Data Mining\Programming Assignment - 1\Data Files\Images'

# List all the available classes (subdirectories) in the directory
all_classes = [d for d in os.listdir(main_dir) if os.path.isdir(os.path.join(main_dir, d))]
print(f"Available Classes: {all_classes}")

# Randomly select two classes from the available classes
selected_classes = random.sample(all_classes, 2)
print(f"Selected Classes for Analysis: {selected_classes}")

# Filter dataset to only the randomly selected classes
filtered_train_files = [file for file, label in zip(train_files, train_labels) if label in selected_classes]
filtered_train_labels = [label for label in train_labels if label in selected_classes]

# Prepare training histograms for the selected classes
filtered_histograms = np.array([reduced_edge_histograms[i] for i, file in enumerate(train_files) if file in filtered_train_files])

# Standardize the data for the selected classes
scaler = StandardScaler()
filtered_histograms_scaled = scaler.fit_transform(filtered_histograms)

# Initialize values for parameter C
C_values = [0.1, 1, 10, 100]

# Dictionary to store the mean accuracies for standard and stratified cross-validation
standard_cv_accuracies = {}
stratified_cv_accuracies = {}

# Set up 5-fold cross-validation objects
kf = KFold(n_splits=5, shuffle=True, random_state=42)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Perform cross-validation for each value of C
for C in C_values:
    # Initialize the classifier with the given C value
    svc = LinearSVC(C=C, max_iter=1000, random_state=42)
    
    # Standard 5-fold cross-validation
    standard_scores = cross_val_score(svc, filtered_histograms_scaled, filtered_train_labels, cv=kf)
    standard_cv_accuracies[C] = standard_scores.mean()
    
    # Stratified 5-fold cross-validation
    stratified_scores = cross_val_score(svc, filtered_histograms_scaled, filtered_train_labels, cv=skf)
    stratified_cv_accuracies[C] = stratified_scores.mean()

# Display the results
print("Standard 5-fold Cross-Validation Mean Accuracies:")
for C, accuracy in standard_cv_accuracies.items():
    print(f"C={C}: Mean Accuracy = {accuracy:.4f}")

print("\nStratified 5-fold Cross-Validation Mean Accuracies:")
for C, accuracy in stratified_cv_accuracies.items():
    print(f"C={C}: Mean Accuracy = {accuracy:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import LinearSVC
from sklearn.model_selection import cross_val_score, StratifiedKFold, KFold
from sklearn.metrics import accuracy_score

# Initialize values for parameter C
C_values = [0.1, 1, 10, 100]

# Set up cross-validation objects
kf = KFold(n_splits=5, shuffle=True, random_state=42)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Dictionary to store training and validation errors for each C
standard_cv_train_errors = []
standard_cv_val_errors = []
stratified_cv_train_errors = []
stratified_cv_val_errors = []

# Loop over C values
for C in C_values:
    # Initialize the classifier with the given C value
    svc = LinearSVC(C=C, max_iter=1000, random_state=42)
    
    # Standard 5-fold cross-validation
    standard_train_errors = []
    standard_val_errors = []
    for train_index, val_index in kf.split(filtered_histograms_scaled):
        # Split data
        X_train, X_val = filtered_histograms_scaled[train_index], filtered_histograms_scaled[val_index]
        y_train, y_val = np.array(filtered_train_labels)[train_index], np.array(filtered_train_labels)[val_index]
        
        # Train model
        svc.fit(X_train, y_train)
        
        # Calculate training and validation errors
        train_error = 1 - accuracy_score(y_train, svc.predict(X_train))
        val_error = 1 - accuracy_score(y_val, svc.predict(X_val))
        
        standard_train_errors.append(train_error)
        standard_val_errors.append(val_error)
    
    # Store the mean errors for standard cross-validation
    standard_cv_train_errors.append(np.mean(standard_train_errors) * 100)
    standard_cv_val_errors.append(np.mean(standard_val_errors) * 100)

    # Stratified 5-fold cross-validation
    stratified_train_errors = []
    stratified_val_errors = []
    for train_index, val_index in skf.split(filtered_histograms_scaled, filtered_train_labels):
        # Split data
        X_train, X_val = filtered_histograms_scaled[train_index], filtered_histograms_scaled[val_index]
        y_train, y_val = np.array(filtered_train_labels)[train_index], np.array(filtered_train_labels)[val_index]
        
        # Train model
        svc.fit(X_train, y_train)
        
        # Calculate training and validation errors
        train_error = 1 - accuracy_score(y_train, svc.predict(X_train))
        val_error = 1 - accuracy_score(y_val, svc.predict(X_val))
        
        stratified_train_errors.append(train_error)
        stratified_val_errors.append(val_error)
    
    # Store the mean errors for stratified cross-validation
    stratified_cv_train_errors.append(np.mean(stratified_train_errors) * 100)
    stratified_cv_val_errors.append(np.mean(stratified_val_errors) * 100)

# Plotting the error curves
plt.figure(figsize=(10, 6))
plt.plot(C_values, standard_cv_train_errors, label="Standard CV Training Error", marker='o')
plt.plot(C_values, standard_cv_val_errors, label="Standard CV Validation Error", marker='o')
plt.plot(C_values, stratified_cv_train_errors, label="Stratified CV Training Error", marker='o')
plt.plot(C_values, stratified_cv_val_errors, label="Stratified CV Validation Error", marker='o')
plt.xscale('log')
plt.xlabel("C (Log Scale)")
plt.ylabel("Mean Error (%)")
plt.title("Training and Validation Errors for Different Values of C")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import accuracy_score

# Identify the C value with the lowest mean validation error from stratified cross-validation
best_C = C_values[np.argmin(stratified_cv_val_errors)]
print(f"Best C value (lowest stratified 5-fold validation error): {best_C}")

# Initialize SVM with the best C value
best_svc = LinearSVC(C=best_C, max_iter=1000, random_state=42)

# Train the SVM classifier on the full training set
best_svc.fit(filtered_histograms_scaled, filtered_train_labels)

# Prepare and standardize the test data
test_labels = [file_to_label[file] for file in test_files if file_to_label[file] in selected_classes]
test_histograms = np.array([reduced_edge_histograms[i] for i, file in enumerate(test_files) if file_to_label[file] in selected_classes])
test_histograms_scaled = scaler.transform(test_histograms)  # Use the same scaler from training

# Predict on the test set
y_test_pred = best_svc.predict(test_histograms_scaled)

# Calculate the test error
test_accuracy = accuracy_score(test_labels, y_test_pred)
test_error = (1 - test_accuracy) * 100  # Convert to percentage

print(f"Test Error for the best C value ({best_C}): {test_error:.2f}%")
```
